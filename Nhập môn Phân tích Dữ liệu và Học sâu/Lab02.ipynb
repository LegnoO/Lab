```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/LegnoO/Lab/main/Nh%E1%BA%ADp%20m%C3%B4n%20Ph%C3%A2n%20t%C3%ADch%20D%E1%BB%AF%20li%E1%BB%87u%20v%C3%A0%20H%E1%BB%8Dc%20s%C3%A2u/Lab1/dulieuxettuyendaihoc2.txt")

print(df)
```

```python
#Phần 1: Thống kê dữ liệu:
#1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
print(df.sort_values(by=['DH1'],ascending=True))
print(df['DH1'].head(10))
```

```python
#2. Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
print(df.sort_values(by='GT',ascending=True))
```

```python
#3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT
print(pd.pivot_table(data=df,index='KT',values='DH1',aggfunc=['count','sum','mean','median','min','max','std']))
```

```python
#4. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT và KV
print(pd.pivot_table(data=df,index=['KT','KV'],values='DH1',aggfunc=['count','sum','mean','median','min','max','std']))
```

```python
#5. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT, KV và DT
print(pd.pivot_table(data=df,index=['KT','KV','DT'],values='DH1',aggfunc=['count','sum','mean','median','min','max','std']))
```

```python
#Phần 2: Trình bày dữ liệu:
#1. Hãy trình bày dữ liệu biến: GT
print(df['GT'])
```

```python
#2. Hãy trình bày dữ liệu lần lượt các biến: US_TBM1, US_TBM2 và US_TBM3
print(df[['US_TBM1','US_TBM2','US_TBM3']])
```
